```python
!pip install control
import matplotlib.pyplot as plt
import control as ctrl

# Definir as funções de transferência G1 e G2
G1 = ctrl.TransferFunction([1], [1, 2, 1])  # tf(1, [1 2 1])
G2 = ctrl.TransferFunction([10], [1, 2, 10])  # tf(10, [1 2 10])

# Multiplicação das funções de transferência Gc = G1 * G2
Gc = G1 * G2

# Plotar a resposta ao degrau de G1, G2 e Gc
t1, y1 = ctrl.step_response(G1)
t2, y2 = ctrl.step_response(G2)
t3, y3 = ctrl.step_response(Gc)

# Plotar todas as respostas ao degrau
plt.plot(t1, y1, label="G1")
plt.plot(t2, y2, label="G2")
plt.plot(t3, y3, label="Gc")

# Configurações do gráfico
plt.title("Respostas ao Degrau")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# Função conv: Convolução para multiplicação dos polinômios
def conv(poly1, poly2):
    return np.convolve(poly1, poly2)

# Definir as funções de transferência G1, G2 e G3
G1 = ctrl.TransferFunction([2], conv([1, 1j+1], [1, -1j+1]))  # tf(2, conv([1 i+1], [1 -i+1]))
G2 = ctrl.TransferFunction([5], conv([1, 2j+1], [1, -2j+1]))  # tf(5, conv([1 2i+1], [1 -2i+1]))
G3 = ctrl.TransferFunction([10], conv([1, 3j+1], [1, -3j+1]))  # tf(10, conv([1 3i+1], [1 -3i+1]))

# Plotar a resposta ao degrau de G1, G2 e G3
t1, y1 = ctrl.step_response(G1)
t2, y2 = ctrl.step_response(G2)
t3, y3 = ctrl.step_response(G3)

# Plotar todas as respostas ao degrau
plt.plot(t1, y1, label="G1")
plt.plot(t2, y2, label="G2")
plt.plot(t3, y3, label="G3")

# Configurações do gráfico
plt.title("Respostas ao Degrau de G1, G2 e G3")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.show()
```
