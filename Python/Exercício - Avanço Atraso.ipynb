```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função de transferência
num = [1, 8]
den = [1, 19, 108, 180]
G = ct.tf(num, den)

# Calculando os polos
poles = ct.poles(G)

# Encontrando o polo dominante
dominant_pole = np.min(poles.real)
print("Polo dominante:", dominant_pole)

# Plotando o diagrama de polos e zeros
plt.figure()
ct.pzmap(G)
plt.grid(True)
plt.xlabel('Real')
plt.ylabel('Imaginário')
plt.title('Diagrama de Polos e Zeros')
plt.show()
```
