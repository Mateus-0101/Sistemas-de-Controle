```python
import matplotlib.pyplot as plt
import numpy as np

# Define a função de transferência
def transferencia(R, R_ldr):
  return (R_ldr / (R + R_ldr))

# Define o valor de R
R = 10000  # 10kΩ

# Define a faixa de resistência do LDR
R_ldr_min = 5000  # 5kΩ
R_ldr_max = 41250  # 41,25kΩ

# Define o passo da resistência do LDR
passo_R_ldr = 5000  # 5kΩ

# Cria um array de resistências do LDR
R_ldr = np.arange(R_ldr_min, R_ldr_max + passo_R_ldr, passo_R_ldr)

# Calcula a tensão de saída para cada resistência do LDR
V_saida = transferencia(R, R_ldr) * 5  # Multiplica por 5 para obter a tensão de saída

# Plota o gráfico
plt.plot(R_ldr, V_saida)
plt.xlabel("Resistência do LDR (Ω)")
plt.ylabel("Tensão de Saída (V)")
plt.title("Gráfico da Tensão de Saída x Resistência do LDR")
plt.grid(True)

# Mostra os valores inicial e final de V
print("Valor inicial de V:", V_saida[0])
print("Valor final de V:", V_saida[-1])

# Exibe o gráfico
plt.show()

```
